```python

import json
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
json_gt = [json.loads(line) for line in open('../datasets/1280x720_original_task_bags 2023_10-2024_01_10_19_13_48/tusimple.json')]
```

```python
for l, gt in enumerate(json_gt[::-1]):
    gt_lanes = gt['lanes']
    y_samples = gt['h_samples']
    
    colors = [(0,0,255), (0,255,0), (255,0,0), (255,255,0), (0,255,255)]
    
    raw_file = gt['raw_file']
    gt_lanes_vis = [[(x,y) for (x,y) in zip(lane, y_samples) if x >= 0] for lane in gt_lanes]
    img = cv.imread(f"../datasets/1280x720_original_task_bags 2023_10-2024_01_10_19_13_48/images/{raw_file}")
    
    for k, lane in enumerate(gt_lanes_vis):
        for pt in lane:
            cv.circle(img, pt, 5, colors[k], -1)
            # cv.line(img, (0, pt[1]), (img.shape[0], pt[1]), (0,255,0), 1)
    cv.imshow(f'img{l}', img)
    # escape to quit, enter to continue
    if cv.waitKey(0) == 13:
        cv.destroyAllWindows()
        continue
    elif cv.waitKey(0) == 27:
        cv.destroyAllWindows()
        break
    
plt.imshow(img)
plt.show()
```
